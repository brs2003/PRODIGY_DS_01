```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#DATA FOR CONSTITUENCY WISE ELIGIBLE ELECTORS

constituency_data = {
    
    "Constituency": ["Kasaragod", "Kannur", "Vatakara", "Wayanad", "Kozhikode", "Malappuram", "Ponnani", "Palakkad",
                     "Alathur (SC)", "Thrissur", "Chalakudy", "Ernakulam", "Idukki", "Kottayam", "Alappuzha", 
                     "Mavelikkara (SC)", "Pathanamthitta", "Kollam", "Attingal", "Thiruvananthapuram"],
    
    "Male": [701475, 646181, 681615, 721054, 691096, 745978, 729255, 682281, 648437, 708317, 634347, 640662, 615084, 
             607502, 674066, 630307, 683307, 631625, 653549, 689155],
    
    "Female": [750741, 712181, 740246, 741354, 738509, 733931, 741522, 715849, 689047, 774718, 676161, 683370, 635064, 
               647306, 726008, 701564, 746384, 695004, 743223, 741317],
    
    "Third Gender": [14, 6, 22, 15, 26, 12, 27, 13, 12, 20, 21, 15, 9, 15, 9, 9, 9, 19, 35, 59]
}
```

```python
# Calculating the total number of electors in each constituency
constituency_data['Total'] = [sum(x) for x in zip(constituency_data["Male"], constituency_data["Female"], constituency_data["Third Gender"])]
constituency_data['Total']
```

```python
# Creating DataFrame
constituency_df = pd.DataFrame(constituency_data)
```

```python
# Plotting the bar chart
fig, ax = plt.subplots(figsize=(14, 8))

bar_width = 0.25
index = range(len(constituency_df))

bar1 = ax.bar(index, constituency_df["Male"], bar_width, label="Male")
bar2 = ax.bar([i + bar_width for i in index], constituency_df["Female"], bar_width, label="Female")
bar3 = ax.bar([i + 2 * bar_width for i in index], constituency_df["Third Gender"], bar_width, label="Third Gender")

ax.set_xlabel("Constituency")
ax.set_ylabel("Number of Electors")
ax.set_title("Distribution of Electors by Constituency and Gender")
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(constituency_df["Constituency"], rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Creating the histogram
plt.figure(figsize=(14, 8))

# Plotting the histogram of total electors
plt.hist(constituency_df['Total'], bins=10, edgecolor='black')

# Adding labels and title to the plot
plt.xlabel("Total Number of Electors")  # X-axis label
plt.ylabel("Number of Constituencies")  # Y-axis label
plt.title("Distribution of Electors Across Constituencies")  # Title of the plot

# Ensuring the plot layout is tight so nothing is cut off
plt.tight_layout()

# Displaying the plot
plt.show()
```
